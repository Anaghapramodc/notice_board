```python
import sqlite3
conn=sqlite3.connect("database.db")
c=conn.cursor()
c.execute("""
     CREATE TABLE customer(
         id INTEGER PRIMERY KEY,
         name TEXT,
         age INTEGER,
         city TEXT
     )
 """)
```
